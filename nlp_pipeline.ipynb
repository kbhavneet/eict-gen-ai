```python
from transformers import pipeline
```

```python
# sentiment-analysis
sentiment=pipeline('sentiment-analysis', model='tabularisai/multilingual-sentiment-analysis')
sentiment('customer service was very good')
```

```python
# NER(Named Entity Recognition)
ner_tagger=pipeline('ner', model='Babelscape/wikineural-multilingual-ner')
entities=ner_tagger('Amitabh Bachchan is my favorite actor and he lives in Mumbai.')
print(entities)
```

```python
# zero-shot-classification
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel', 'cooking', 'dancing']
classifier(sequence_to_classify, candidate_labels)
```

```python
# zero-shot-classification
regulation_forecast='The report predicts that stricter data regulations will force companies to redesign their AI pipelines and improve model transparency.'
issues_classifier=pipeline('zero-shot-classification')
labels=["policy_change", "technology_trend", "business_strategy"]
issues_result=issues_classifier(regulation_forecast, labels)
print(issues_result)
```

```python
#question answering
qa=pipeline('question-answering')
qa({
    "context": "Rohit Sharma lives in Mumbai and is a great cricketer",
    "question": "Who is Rohit Sharma?"
})
```

```python
#question answering
qa=pipeline('question-answering', model='consciousAI/question-answering-roberta-base-s-v2')
qa({
    "context": "Rohit Sharma lives in Mumbai and is a great cricketer",
    "question": "Where does Rohit Sharma live?"
})
```

```python
# text generation
gen=pipeline('text-generation')
gen('AI will transform the businesses because ', max_length=50)
```

```python
# My own practice
# load the zephyr model for text generation
# Zephyr models are lightweight 7-billion-parameter language models that have been trained to be more conversational and follow instructions well.
generator = pipeline(
    "text-generation",
    model="alignment-handbook/zephyr-7b-sft-full",
    dtype="auto",
    device_map="auto"   # automatically uses GPU if available
)

prompt = "Explain the importance of model alignment in simple terms."

output = generator(
    prompt,
    max_length=200,
    do_sample=True,
    top_p=0.9,
    temperature=0.7
)

print(output[0]["generated_text"])
```

```python
from sentence_transformers import SentenceTransformer,util
```

```python
#similarity searching
model=SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
sent1='ML is amazing'
sent2='AI is awesome'
sent3='I have to go to IIT Kanpur'
emb1=model.encode(sent1,convert_to_tensor=True)
emb2=model.encode(sent3,convert_to_tensor=True)
score=util.pytorch_cos_sim(emb1,emb2)
print(emb1)
print(score)
```
